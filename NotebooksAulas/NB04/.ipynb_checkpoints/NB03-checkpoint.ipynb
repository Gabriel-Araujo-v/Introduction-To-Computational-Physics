```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from itertools import product
```

```python

# Número total de partículas no sistema
N = 64  # Você pode mudar este número para simular diferentes tamanhos de sistema

# Número inicial de partículas no lado esquerdo (todas as partículas começam no lado esquerdo)
n = N

# Número total de passos de tempo para simular
max_time = 100*N  # Você pode mudar isso para simular tempos mais longos ou mais curtos

# Lista para registrar o número de partículas no lado esquerdo ao longo do tempo
n_left = []

# Loop de simulação ao longo do tempo
for t in range(max_time):
    # Registrar o número atual de partículas no lado esquerdo
    n_left.append(n)
    
    # Gerar um número aleatório entre 0 e 1
    # Este número determinará a direção do movimento da partícula
    r = np.random.rand()
    
    # Calcular a probabilidade de uma partícula mover-se da esquerda para a direita
    # Isto é proporcional à razão de partículas no lado esquerdo (n/N)
    p_left_to_right = n / N
    
    # Decidir a direção do movimento da partícula com base no número aleatório
    if r < p_left_to_right:
        # A partícula move-se da esquerda para a direita
        # Garantir que há partículas no lado esquerdo para mover
        if n > 0:
            n -= 1  # Diminuir o número de partículas no lado esquerdo em 1
    else:
        # A partícula move-se da direita para a esquerda
        # Garantir que há partículas no lado direito para mover
        if N - n > 0:
            n += 1  # Aumentar o número de partículas no lado esquerdo em 1

# Após a simulação, plotar os resultados para visualizar a evolução do sistema
plt.figure(figsize=(5,3))  # Definir o tamanho da figura para melhor visibilidade

# Plotar o número de partículas no lado esquerdo ao longo do tempo
plt.plot(np.asarray(n_left)/N, label='Número de partículas no lado esquerdo')

# Rotular os eixos e adicionar um título
plt.xlabel('Passos de tempo')  # Rótulo do eixo X
plt.ylabel('# de partículas no lado esquerdo')  # Rótulo do eixo Y
plt.title('Evolução das partículas entre as duas metades de uma caixa')  # Título do gráfico

# Adicionar uma legenda ao gráfico
plt.legend()

# Adicionar uma grade para facilitar a leitura do gráfico
plt.grid(True)

# Exibir o gráfico
plt.show()
```

### Problem 7.3: Equilibrium Fluctuations

**(a)** Modify the program so that averages are taken after equilibrium has been reached. What is the maximum deviation of $n(t)$ from $N/2$ for $N = 64, 400, 800,$ and $3600$? Run for a time that is long enough to yield meaningful results. How do you know if your results are meaningful? How do your results for the maximum deviation depend on $N$?

**(b)** A measure of the equilibrium fluctuations is the variance $\sigma^2$ defined as
   $$
   \sigma^2 = \langle n^2 \rangle - \langle n \rangle^2. \tag{7.1}
   $$
The relative magnitude of the fluctuations is $\sigma / \langle n\rangle$. Compute the variance of $n$ for the same values of $N$ considered in part (a). How do the relative fluctuations, $\sigma / \langle n\rangle$, depend on $N$?


```python
# Lista dos números totais de partículas N a serem simulados
N_values = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096]

# Número total de passos de tempo para simular
max_time = 10000  # Aumentado para garantir resultados significativos

# Tempo de equilíbrio após o qual começamos a coletar dados
equilibration_time = max_time // 2  # Usamos metade do tempo total para equilíbrio

# Inicializar listas para armazenar os resultados para cada N
max_deviations = []
relative_fluctuations = []

# Loop sobre diferentes valores de N
for N in N_values:
    # Inicializar o número de partículas no lado esquerdo
    n = N
    
    # Lista para registrar n(t) após o equilíbrio
    n_left = []
    
    # Loop de simulação ao longo do tempo
    for t in range(N*max_time):
        # Gerar um número aleatório entre 0 e 1
        r = np.random.rand()
        
        # Calcular a probabilidade de uma partícula mover-se da esquerda para a direita
        p_left_to_right = n / N
        
        # Decidir a direção do movimento da partícula
        if r < p_left_to_right:
            # Partícula move-se da esquerda para a direita
            if n > 0:
                n -= 1
        else:
            # Partícula move-se da direita para a esquerda
            if n < N:
                n += 1
        
        # Após o tempo de equilíbrio, registrar n(t)
        if t >= equilibration_time:
            n_left.append(n)
    
    # Converter a lista em um array numpy para cálculos
    n_left = np.array(n_left)
    
    # Calcular a máxima desvio de n(t) em relação a N/2
    max_deviation = np.max(np.abs(n_left - N/2))
    max_deviations.append(max_deviation)
    
    # Calcular a média de n(t)
    n_avg = np.mean(n_left)
    
    # Calcular a média de n(t)^2
    n2_avg = np.mean(n_left**2)
    
    # Calcular a variância sigma^2
    variance = n2_avg - n_avg**2
    
    # Calcular o desvio padrão sigma
    sigma = np.sqrt(variance)
    
    # Calcular a flutuação relativa sigma / <n>
    relative_fluctuation = sigma / n_avg
    relative_fluctuations.append(relative_fluctuation)
    
    # Imprimir os resultados para este N
    print(f"N = {N}")
    print(f"Máximo desvio de n(t) em relação a N/2: {max_deviation}")
    print(f"Média do número de partículas no lado esquerdo <n>: {n_avg}")
    print(f"Variância sigma^2: {variance}")
    print(f"Flutuação relativa sigma/<n>: {relative_fluctuation}")
    print("-" * 40)

# Plotar o máximo desvio vs N
plt.figure(figsize=(5,3))
plt.plot(N_values, max_deviations, 'o-')
plt.xlabel('Número total de partículas N')
plt.ylabel('Máximo desvio de n(t) em relação a N/2')
plt.title('Máximo desvio vs N')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()

# Plotar as flutuações relativas vs N
plt.figure(figsize=(5,3))
plt.plot(N_values, relative_fluctuations, 'o-')
plt.xlabel('Número total de partículas N')
plt.ylabel('Flutuação relativa sigma / <n>')
plt.title('Flutuação relativa vs N')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```
